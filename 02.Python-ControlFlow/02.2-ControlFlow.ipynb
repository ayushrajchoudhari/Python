```python
## for loop
## start(included)(default = 0), end before (excluded)
for i in range(5):
    print(i)
```

```python
## start (included) , end before (excluded)
for i in range(1, 5):
    print (i)
```

```python
## start, end before, steps (here increasing by 2)
for i in range(1, 5, 2):
    print(i)
```

```python
## string
str = "Hello World!"

for i in str:
    print(i)
```

```python
## while loop
count = 0

while count < 5: 
    print (count)
    count += 1
```

### Loop Control Statements

```python
## break - exits the loop prematurely
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue - skips the iteration and starts the next iteration
for i in range(10):
    if i==5:
        continue
    print(i)
```

```python
## pass - is a null operation and does nothing
for i in range(10):
    if i==5:
        print("the number is 5")
        pass
    print(i)
```

```python
## Nested Loops - loop within a loop
for i in range(3):
    for j in range(3):
        print(f"i:{i} and j:{j}")
```

```python
## Calculate the sum first first n natural numbers using while and for loop

## while loop
n = int(input("Enter the n: "))
count = 0
sum = 0

while count < n+1:
    sum += count
    count+=1
print(sum)

## for loop
sum = 0
for i in range(n+1):
    sum += i
print(sum)
```

```python
## Print prime numbers between 1 to 100
for i in range(1,101):
    if i > 1:
        for j in range(2, i):
            if i%j==0:
                break
        else:   # else can be written along with for loop as well
                print(i)
```

### Conclusion
Loops are powerful constructs in Python that allows you to execute a block of code multiple times. By understanding and using for and while loops, along with loop control statements like break, continue, and pass, you can handle a wide range of programming tasks efficiently.
