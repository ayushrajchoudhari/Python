```python
## if statement
## executes when if statement conditin evaluates to true.
if(True):
    print("Condition is true -> if block will execute.")
```

```python
## else statement
## executes when if statement condition evaluates to false.
if(False):
    print("Condition is false -> if block will not execute.")
else :
    print("if condition is false -> else block will execute.")
```

```python
## elif
## allows to check multiple conditions at once.
## executes when all previous if and elif statement conditons evaluates to false and the current elif statement condition evaluates to true.
## when if and all elif statement conditions evaluates to false the else statement at the end executes.
if(False):
    print("Condition is false -> if block will not execute.")
elif(True):
    print("Conditin is true -> elif block will execute.")
else : 
    print("All conditions are false -> else block will execute.")
```

```python
## Nested Conditions
## condition within another condition, can be multiple level deep.
if(True):
    print("1st if block.")

    if(True):
        print("2nd if block.")

    else: 
        print("2nd else block.")
        
else: 
    ("1st else block.")
```

```python
## Practical Example
num = 4

if(num == 0):
    print("Number is zero!")

elif(num > 0):
    print("Number is positive!")

    if(num % 2 == 0):
        print("Number is even!")
    else : 
        print("Number is odd!")

else : 
    print("Number is negative")
```

```python
## Determine if a year is a leap yeat using nested condition statement
year = int(input("Enter the year: "))

if(year % 4 == 0):
    if(year % 100 == 0):
        if(year % 400 == 0):
            print("Year is a leap year!")
        else : 
            print("Year is not a leap year!")
    else:
        print("Year is a leap year!")
else : 
    print("Year is not a leap year!")
```

```python
## Basic Calculator
num1 = int(input("Enter number1: "))
num2 = int(input("Enter number2: "))
operation = str(input("Enter an operator from : + - * /"))

if(operation == '+'):
    print (num1+num2)
elif(operation == '-'):
    print(num1-num2)
elif(operation == '*'):
    print(num1*num2)
elif(operation == '/'):
    print(num1/num2)
else:
    print("Operation choice is invalid!")
```
