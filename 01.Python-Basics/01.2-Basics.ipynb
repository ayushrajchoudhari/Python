### Variable
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In python, variables are created when you assign a value to them, and they do not need explicit declaration to reserv memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring And Assigning Variable
age = 22
height = 5.11
name = "Ayush"
is_male = True

## Printing the variables
print("Age: " , age )
print("Height: ", height)
print("Name: ", name)
print("Is Male: ", is_male)
```

```python
## Naming Conventions
## Variable names should be descriptive
## They must start with a letter or an '_' and contains letters, numbers and underscores
## variabe names are case sensitive

## Valid variable names
first_name = "Ayush"
last_name = "Raj"
```

```python
## Invalid variable names
# 1name = "Hello"
# first-name = "World"
# $last-name = "Kumar"
# @name = "Unknown"
```

```python
## Case sensitivity
name = "Ayush"
Name = "Raj"

print(name)
print(Name)
```

```python
## Understanding Variable Types
## Python is dynamically typed, type of a variable is determined at runtime
age = 22    # int
height = 5.11   # float
name = "Ayush"  # str
is_male = True  # bool
```

```python
print(type(age))
print(type(height))
print(type(name))
print(type(is_male))
```

```python
## Type Conversion
age_str = str(age)
print(age_str)
print(type(age_str))
```

```python
age = '25'
print(age)
print(type(age))
print(type(int(age)))
```

```python
## Dynamic Typing
## Python allows the type of a variable to change as the program executes
var = 10
print(var, type(var))

var = "Hello"
print(var, type(var))

var = 4.5
print(var, type(var))
```

```python
## Simple Calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
quotient = num1 / num2

print("Sum: ", sum)
print("Difference: ", difference)
print("Product: ", product)
print("Quotient: ", quotient)
```
