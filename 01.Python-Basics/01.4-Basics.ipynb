### Arithmetic Operator

```python
a = 10
b = 15

## addition +
add_result = a + b
print(add_result)

## substraction -
sub_result = b - a
print(sub_result)

## multiplication *
mult_result = a * b
print(mult_result)

## division /
div_result = b / a
print(div_result)

## Floor Division // - removes decimal ponts
floor_div_result = b // a
print(floor_div_result)

## Modulus %
mod_result = b % a
print(mod_result)

## Exponential **
exp_result = a ** b
print(exp_result)
```

### Comparison Operator

```python
## Greater Than >
a = 10 
b = 5

a > b
```

```python
## Less Than <
a = 15
b = 20

a < b
```

```python
## Greater Than or Equal To >=
a = 15
b = 15

a >= b
```

```python
## Less Than or Equal To <=
a = 20
b = 20

a <= b
```

### Logical Operator

```python
## and 
a = True
b = False
c = True
d = False

print(a and b)
print(a and c)
print(b and d)
```

```python
## or
a = True
b = False
c = True
d = False

print(a or b)
print(a or c)
print(b or d)
```

```python
## not
a = True
b = False

print(not a)
print(not b)
```
