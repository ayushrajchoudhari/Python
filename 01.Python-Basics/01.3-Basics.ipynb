### Data Types
Definition:
    * Data types are classification of data which tells the compiler or interpreter how the programmer intends to use the data.
    * They determine the type of operations that can be performed on the data, the values that the data can take, and the amount of memory needed to store the data.

Importance of Data Type in Programming:
    * Data types ensures that data is stored in an efficient way.
    * They help in performing correct operation on data.
    * Proper use of data types can prevent errors and bugs in the program.
